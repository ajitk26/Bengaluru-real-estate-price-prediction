```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
data3[~data3['total_sqft'].apply(is_float)].head(15)
```

```python
def convert_sqft(x):
    num=x.split('-')
    if len(num)==2:
        return (float(num[0]) + float(num[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft('3090 - 5002')
```

```python
data4=data3.copy()
data4['total_sqft']=data4['total_sqft'].apply(convert_sqft)
data4.head(10)
```

```python
data5=data4.copy()

data5['price_per_sqft']=data5['price']*100000/data5['total_sqft']
data5.head()
```

```python
data5.location=data5.location.apply(lambda x: x.strip())
location_stats=data5.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_less_than_10=location_stats[location_stats<=10]
location_less_than_10
```

```python
data5.location=data5.location.apply(lambda x: 'other' if x in location_less_than_10 else x)
len(data5.location.unique())
```

```python
data5.head(10)
```

```python
data5[data5.total_sqft/data5.bhk<300].head()# drop these data. 
```

```python
data6=data5[~(data5.total_sqft/data5.bhk<300)]
data6.shape
```

```python
data6.price_per_sqft.describe()
```

```python
# Remove the price_per_sqft outliers

def remove_pps_outliers(df):
    data_out=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft )
        sd=np.std(subdf.price_per_sqft)
        reduced_data=subdf[(subdf.price_per_sqft>(m-sd)) & (subdf.price_per_sqft<=(m+sd))]
        data_out=pd.concat([data_out,reduced_data], ignore_index=True)
    return data_out


data7=remove_pps_outliers(data6)
data7.shape
```

```python
def plot_scatter(df,location):
    bhk2=df[(df.location==location) & (df.bhk==2)]
    bhk3=df[(df.location==location) & (df.bhk==3)]
    
    plt.figure(figsize=(12,8))
    plt.scatter(bhk2.total_sqft, bhk2.price, color='blue', label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft, bhk3.price, color='green', label='3 BHK', s=50)
    plt.xlabel("Total square feet area")
    plt.ylabel("price")
    plt.title(location)
    plt.legend()
    
plot_scatter(data7,"Rajaji Nagar")
```

```python
# # As we can see in the above figure that price of some  3bhk apartment is less than the 2bhk apartment 
# with the same sq. ft. area

# Now we can remove those 2bhk apartment whose whose price_per_sqft is more than 3bhk apartment

def remove_bhk_outlier(df):
    exclude_indices=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
            
data8=remove_bhk_outlier(data7)
data8.shape
    
    
```

```python
# we need to remove the data whose no. bath is greater than no. of bedrooms

data9=data8[data8.bath<data8.bhk+2]
data9
```

```python
# Now prepare for the final data and remove the unnecessay data from the dataset
data10=data9.drop(['size', 'price_per_sqft'], axis='columns')
data10.head(10)
```

```python
# Now use the one-hot encoding that is used to convert categorical variable into the dummies/indicator variables

dummies=pd.get_dummies(data10.location)
dummies.head(3)
```

```python
data11=pd.concat([data10,dummies.drop('other', axis='columns')], axis='columns')
data11.head(5)
```

```python
data12=data11.drop('location', axis='columns')
data12.head(5)
```

```python
data12.shape
```

```python
X=data12.drop('price', axis='columns')
X.head(4)
```

```python
y=data12.price
y.head(4)
```

```python
#now train the model

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
def pric_predict(location,sqft,bath,bhk):
    loc_index=np.where(X.columns==location)[0][0]
    
    x=np.zeros(len(X.columns))
    x[0]=sqft
    x[1]=bath
    x[2]=bhk
    if loc_index>=0:
        x[loc_index]=1
        
    return lr.predict([x])[0]
```

```python
pric_predict('1st Phase JP Nagar',1875,2,3)
```

```python
pric_predict('Indira Nagar',1000,2,2)
```
